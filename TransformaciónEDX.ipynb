```python
#Importamos pandas:
import pandas as pd
```

```python
#Importamos el Dataset:
edx = pd.read_csv("Datasets/edx_courses.csv")
```

```python
edx.head(5)
```

```python
#Para agregar una columna con el precio del certificado en formato numerico:
edx["certificate for(USD)"]=0

#Para completar los datos de la nueva columna:
for i in range(len(edx)):
    if len(edx["price"][i])==43:
        edx.loc[i, "certificate for(USD)"] = edx["price"][i][37:39]
    elif len(edx["price"][i])==44:
        edx.loc[i, "certificate for(USD)"] = edx["price"][i][37:40]
    elif len(edx["price"][i])==42:
        edx.loc[i, "certificate for(USD)"] = edx["price"][i][37:38]

#Para eliminar la columna price:
edx.drop("price", axis=1, inplace=True)
```

```python
edx.head(5)
```

```python
#Para obtener los valores numericos de course_length:
for i in range(len(edx)):
    if len(edx.loc[i, "course_length"]) == 7:
        edx.loc[i, "course_length"] = edx.loc[i,"course_length"][0]
    elif len(edx.loc[i, "course_length"]) == 8:
        edx.loc[i, "course_length"] = edx.loc[i,"course_length"][:2]

#Para renombrar la columna:
edx.rename(columns={"course_length": "course_length(weeks)"}, inplace=True)
```

```python
#Para obtener la cadena de valores del tipo numero-numero de course_effort:
for i in range(len(edx)):
    if len(edx.loc[i, "course_effort"]) == 18:
        edx.loc[i, "course_effort"] = edx.loc[1, "course_effort"][:3]
    elif len(edx.loc[i, "course_effort"]) == 19:
        edx.loc[i, "course_effort"] = edx.loc[i,"course_effort"][:4]
    elif len(edx.loc[i, "course_effort"]) == 20:
        edx.loc[i, "course_effort"] = edx.loc[i,"course_effort"][:5]

#Para obtener cada valor en dos nuevas listas:
valores1 = []
valores2 = []

for i in range(len(edx)):
    valores = edx.loc[i, "course_effort"].split('–')
    valores1.append((valores[0]))
    valores2.append((valores[1]))

#Para obtener el promedio de cada valor y asignarselo a course_effort
for i in range(len(edx)):
    edx.loc[i,"course_effort"] = (int(valores1[i])+int(valores2[i]))/2

#Para renombrar la columna:
edx.rename(columns={"course_effort": "course_effort(hours_per_week)"}, inplace=True)
```

```python
#Para eliminar las comas de la columna n_enrolled:
edx["n_enrolled"] = edx["n_enrolled"].str.replace(",","")
```

```python
#Para ver la cantidad de nulos de la columna n_enrolled:
edx.info()
```

```python
#Para eliminar aquellas filas que no tienen datos en n_enrolled:
#Esta columna sera muy importante en nuestro analisis, por eso eliminamos aquellas filas que tengan datos nulos.
edx = edx.dropna(subset=['n_enrolled'])
```

```python
#Para Transformar los valores de n_enrolled a enteros:
#Para poder trabajar con ellos.

edx['n_enrolled'] = edx['n_enrolled'].astype(int)
```

```python
#Para obtener el total de inscriptos:
total = edx['n_enrolled'].sum()

#Para obtener el porcentaje de cada curso sobre el total:
edx['enrolled_percent'] = edx['n_enrolled'].apply(lambda x: (x / total) * 100)
```

```python
edx.sort_values("enrolled_percent", ascending=False, inplace=True)
```

```python
edx.head(5)
```

```python
#Para hacer la sumatoria acumulada de enrolled percent:
edx['enrolled_acumulado'] = edx['enrolled_percent'].cumsum()
```

```python
#Para resetear el indice con el orden actual:
edx = edx.reset_index(drop=True)

# Crear la nueva columna dividiendo en 10 categorías basadas en el porcentaje acumulado
edx['segmento'] = 0

#Para completar los datos de segmento:
#Cada segmento represetara al 10% del total de subscriptos.
#El primer segmento lo ocuparan aquellos cursos que tienen mas inscriptos, hasta el 10% del total, y asi.
for i in range(len(edx)):
    if edx.iloc[i]["enrolled_acumulado"]<=10.5:
        edx.loc[i,"segmento"]=1
    elif edx.iloc[i]["enrolled_acumulado"]<=20.5:
        edx.loc[i,"segmento"]=2
    elif edx.iloc[i]["enrolled_acumulado"]<=30.5:
        edx.loc[i,"segmento"]=3
    elif edx.iloc[i]["enrolled_acumulado"]<=40.5:
        edx.loc[i,"segmento"]=4
    elif edx.iloc[i]["enrolled_acumulado"]<=50.5:
        edx.loc[i,"segmento"]=5
    elif edx.iloc[i]["enrolled_acumulado"]<=60.5:
        edx.loc[i,"segmento"]=6
    elif edx.iloc[i]["enrolled_acumulado"]<=70.5:
        edx.loc[i,"segmento"]=7
    elif edx.iloc[i]["enrolled_acumulado"]<=80.5:
        edx.loc[i,"segmento"]=8
    elif edx.iloc[i]["enrolled_acumulado"]<=90.5:
        edx.loc[i,"segmento"]=9
    else:
        edx.loc[i,"segmento"]=10
```

```python
#Para borrar la columna enrolled_acumulado:
edx = edx.drop("enrolled_acumulado", axis=1)
```

```python
#Para redondear los valores de enrolled_percent:
edx["enrolled_percent"] = edx["enrolled_percent"].round(3)
```

```python
#Para transformar los valores de course_length a float:
edx["course_length(weeks)"] = edx["course_length(weeks)"].astype(float)
```

```python
#Para combinar course_length con course_effort:
for i in range(len(edx)):
    edx.loc[i,"course_effort(total_hours)"]=(edx.loc[i, "course_length(weeks)"])*(edx.loc[i, "course_effort(hours_per_week)"])
```
