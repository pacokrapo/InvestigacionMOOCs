```python
#Importamos pandas:
import pandas as pd
```

```python
#Importamos Dataset:
Udemy = pd.read_csv("Datasets/udemy_courses.csv")
```

```python
#Para ordenar los datos en base al numero de suscriptores
Udemy = Udemy.sort_values("num_subscribers", ascending=False)
```

```python
#Para obtener el total de suscriptos:
total = Udemy['num_subscribers'].sum()

#Para obtener el porcentaje de suscriptos de cada curso sobre el total:
Udemy['subscribers_percent'] = Udemy['num_subscribers'].apply(lambda x: (x / total) * 100)
```

```python
#Para hacer la sumatoria acumulada de subscribers_percent:
Udemy['subscribers_acumulado'] = Udemy['subscribers_percent'].cumsum()
```

```python
Udemy.head(5)
```

```python
#Para resetear el indice con el orden actual:
Udemy = Udemy.reset_index(drop=True)

# Para crear una nueva columna dividiendo en 10 categor√≠as basadas en el porcentaje acumulado:
Udemy['segmento'] = 0

#Para completar los datos de segmento:
#Cada segmento represetara al 10% del total de subscriptos.
#El primer segmento lo ocuparan aquellos cursos que tienen mas inscriptos, hasta el 10% del total, y asi.
for i in range(len(Udemy)):
    if Udemy.iloc[i]["subscribers_acumulado"]<=10.5:
        Udemy.loc[i,"segmento"]=1
    elif Udemy.iloc[i]["subscribers_acumulado"]<=20.5:
        Udemy.loc[i,"segmento"]=2
    elif Udemy.iloc[i]["subscribers_acumulado"]<=30.5:
        Udemy.loc[i,"segmento"]=3
    elif Udemy.iloc[i]["subscribers_acumulado"]<=40.5:
        Udemy.loc[i,"segmento"]=4
    elif Udemy.iloc[i]["subscribers_acumulado"]<=50.5:
        Udemy.loc[i,"segmento"]=5
    elif Udemy.iloc[i]["subscribers_acumulado"]<=60.5:
        Udemy.loc[i,"segmento"]=6
    elif Udemy.iloc[i]["subscribers_acumulado"]<=70.5:
        Udemy.loc[i,"segmento"]=7
    elif Udemy.iloc[i]["subscribers_acumulado"]<=80.5:
        Udemy.loc[i,"segmento"]=8
    elif Udemy.iloc[i]["subscribers_acumulado"]<=90.5:
        Udemy.loc[i,"segmento"]=9
    else:
        Udemy.loc[i,"segmento"]=10
```

```python
#Para obtener el numero de reviews sobre suscriptos:
Udemy['reviews/subscribers'] = Udemy['num_reviews']/Udemy["num_subscribers"]
```

```python
#Para eliminar la columna subscribers_acumulado:
Udemy = Udemy.drop("subscribers_acumulado", axis=1)
```

```python
#Para redondear los valores de subscribers_percent y reviews/subscribers:
Udemy["subscribers_percent"] = Udemy["subscribers_percent"].round(3)
Udemy["reviews/subscribers"] = Udemy["reviews/subscribers"].round(5)
```
