```python
#Importamos pandas:
import pandas as pd
```

Importamos los datos:

```python
#Importamos Datasets:
CourseraCursos = pd.read_csv("Datasets/Coursera_courses.csv")
CourseraReviews = pd.read_csv("Datasets/Coursera_reviews.csv")
```

```python
#Para eliminar filas duplicadas en CourseraReviews:
CourseraReviews = CourseraReviews.drop_duplicates()
```

```python
#Para crear un nuevo dataset con la informacion relevante:
CourseraRating = CourseraReviews[["rating", "course_id"]]
```

```python
# Para calcular los promedios de rating para cada valor único de course_id:
promedios = CourseraRating.groupby("course_id")["rating"].mean()

#Para obtener la cantidad de votos que tiene cada valor unico de course_id:
conteo_votos = CourseraRating["course_id"].value_counts()

#Para ordenar alfabeticamente y que coincidan los valores con los promedios:
conteo_votos = conteo_votos.sort_index()

# Para generar un nuevo DataFrame:
CourseraRatingPromedio = pd.DataFrame({"course_id": promedios.index, "Rating Promedio": promedios.values, "Cantidad Votos": conteo_votos})
```

```python
#Para redondear los valores de Rating Promedio:
CourseraRatingPromedio["Rating Promedio"] = CourseraRatingPromedio["Rating Promedio"].round(2)
```

```python
#Para unir el primer DataFrame con informacion de los cursos con el nuevo generado con informacion de los rating:
Coursera = pd.merge(CourseraCursos, CourseraRatingPromedio, on='course_id')
```
